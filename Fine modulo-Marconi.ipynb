## Esercizio 2

```python
election = px.data.election()
```

```python
election.head()
```

```python
Coderre = election['Coderre']
Bergeron = election['Bergeron']
Joly = election['Joly']
```

```python
Coderre.sum()
```

```python
plt.bar(('Coderre','Bergeron','Joly'),(Coderre.sum(),Bergeron.sum(),Joly.sum()))
plt.title('Result elections')
plt.xlabel('Cadidates')
plt.ylabel('Votes')
plt.show()
```

```python
election4 = election[:4]
election4.plot.bar(x='district',y=['Coderre','Bergeron','Joly'],title='Results of elections')
plt.savefig('Result of elections',dpi=600)
plt.show()
```

```python
election4.plot.bar(x='district',y=['Coderre','Bergeron','Joly'],title='Results of elections',stacked=True)
plt.savefig('Result of elections (stacked)',dpi=600)
plt.show()
```

### Esercizio 2/3

```python
sns.countplot(data=titanic,x='class')
plt.title("Numero di passeggeri per ogni classe")
```

```python
sns.countplot(data=titanic,x='alive')
plt.title("Quantit di sopravvissuti")
```

```python
sns.displot(titanic.fare,kde=True)
plt.title('Distribuzione delle tariffe')
```

```python
sns.boxplot(data=titanic,x="class",y="age")
plt.title("Distribuzione dell'et rispetto alla classe")
```

```python
sns.swarmplot(data=titanic,x="class",y="age")
plt.title("Distribuzione dell'et rispetto alla classe")
```

```python
sns.kdeplot(titanic.age)
plt.title("Distribuzione dell'et")
```

```python
sns.boxplot(data=titanic,x="survived",y="fare")
plt.title("Relazione tra tariffa e passeggeri sopravvisuti")
```

```python
sns.lmplot(data=titanic,x="fare",y="survived")
plt.title("Relazione tra tariffa e passeggeri sopravvisuti")
```

#### Da questi due grafici si capisce che i passeggeri che hanno pagato una tariffa pi alta, hanno avuto una probabilit moderatamente pi elevata rispetto coloro che hanno comprato un biglietto meno caro. Si vede anche che la relazione tra tariffa e sopravvivenza non  lineare. 

### Esercizio 3/3

```python
fig = px.scatter(titanic,x="age",y="deck",color="survived",size="fare",symbol="sex")
fig.update_yaxes(categoryorder="category ascending")
```

#### Dal grafico possiamo concludere che i passeggeri che hanno avuto pi probabilit si sopravvivenza sono queli che si trovavano sui ponti B e C che, allo stesso tempo corrispondevano coi settori con tariffe pi alte.
